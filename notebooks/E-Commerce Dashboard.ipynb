```python
# Cell 10
print("=== NULL VALUES ANALYSIS ===\n")

null_df = pd.DataFrame({
    'Column': df.columns,
    'Null Count': df.isnull().sum().values,
    'Null %': (df.isnull().sum().values / len(df) * 100).round(2)
})

# Sirf jinmein nulls hain woh dikhao
null_df = null_df[null_df['Null Count'] > 0]
null_df = null_df.sort_values('Null %', ascending=False)

print(null_df.to_string(index=False))
```

```python
# Cell 11
duplicates = df.duplicated().sum()
print(f"Duplicate rows: {duplicates}")
print(f"Percentage   : {duplicates/len(df)*100:.2f}%")

# duplicated() — True/False har row ke liye
# .sum() — True = 1, False = 0, so total duplicates count ho jaata hai
```

```python
# Cell 12
categorical_cols = ['Status', 'Fulfilment', 'Sales Channel',
                    'ship-service-level', 'Category', 'B2B']

print("=== UNIQUE VALUES IN CATEGORICAL COLUMNS ===\n")
for col in categorical_cols:
    print(f"📌 {col}:")
    print(f"   {df[col].unique()}")
    print(f"   Total unique: {df[col].nunique()}\n")
```

```python
# Cell 13 — Summary print karo
print("=" * 50)
print("AMAZON SALE REPORT — DATASET SUMMARY")
print("=" * 50)
print(f"\n📦 Total Orders  : {df['Order ID'].nunique():,}")
print(f"📅 Date Range    : {df['Date'].min()} to {df['Date'].max()}")
print(f"💰 Total Revenue : ₹{df['Amount'].sum():,.2f}")
print(f"📊 Categories    : {df['Category'].nunique()}")
print(f"🗺️  States        : {df['ship-state'].nunique()}")
print(f"🏙️  Cities        : {df['ship-city'].nunique()}")
print(f"📋 Order Status  : {df['Status'].unique()}")
print("=" * 50)
```
